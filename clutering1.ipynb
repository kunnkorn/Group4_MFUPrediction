```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel('StudentApplications.xlsx')
```

```python
df.head()
```

```python
#drop columns
df = df.drop(columns = ['LevelID','LevelName','LevelNameEng','ApplicationDate','AcademicSemester','SchoolRegion','SchoolRegionNameEng','ReligionName','MajorName','EntryTypeID' , 'EntryTypeName','FacultyID', 'DepartmentCode','PrefixName','EntryGroupID'])
df
```

```python
#drop columns
df = df.drop(columns = ['HomeRegion','Country','StudentTH','SchoolRegionName','ProvinceNameEng','NationName','SchoolProvince'])
df
```

```python
data = df[['AcademicYear','Sex','FacultyName','EntryGroupName','EntryGPA','StudentType','Status']]
```

```python
data
```

```python
#get only MFU student only
mfu = data[data['Status'] == 'MFU student']

#get Thai student only
mfu_th = mfu[mfu['StudentType'] == 'THAI']

#get academic for each 2559 and 2560
mfu_th_y60 = mfu_th[mfu_th['AcademicYear'] == 2560]

mfu_th_y60
```

```python
mfu_th_y60 = mfu_th_y60.drop(columns = ['AcademicYear','StudentType','Status'])
mfu_th_y60
```

```python
mfu_th_y60[mfu_th_y60.columns[mfu_th_y60.isnull().any()]].isnull().sum() / len(mfu_th_y60) *100
```

```python
mfu_th_y60['EntryGPA'].fillna(mfu_th_y60['EntryGPA'].median(),inplace = True)
mfu_th_y60
```

```python
#get data for each group
numeric_data = mfu_th_y60.select_dtypes(include = 'number')
nominal_data = mfu_th_y60.select_dtypes(exclude = 'number')
```

```python
numeric_data
```

```python
nominal_data
```

```python
#one-hot data
nominal_data = pd.get_dummies(nominal_data)
nominal_data
```

```python
#combine
dff = pd.concat([nominal_data,numeric_data],axis = 1,sort = False)
dff
```

```python
# Rescaling features into a range of [0,1]
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
X = mms.fit_transform(dff)
X = pd.DataFrame(X, columns=dff.columns) 
X
```

```python
# Measure the quality of clusterings with different numbers of clusters using the inertia. 
# For each of the given values of k, perform the following steps:
#     Create a KMeans instance called model with k clusters.
#     Fit the model to the data.
#     Append the value of the inertia_ attribute of model to the list inertias.

from sklearn.cluster import KMeans

ks = range(1, 11) # for 1 to 10 clusters
sse = []

for k in ks:
    # Create a KMeans instance with k clusters: model
    #kmeans = KMeans(n_clusters=k)
    kmeans = KMeans(n_clusters = k)

    # Fit model to samples
    kmeans.fit(X)

    # Append the inertia to the list of inertias
    sse.append(kmeans.inertia_)
```

```python
# Plot the inertia to see which number of clusters is best. Remember: lower numbers are better!
import matplotlib.pyplot as plt

# Plot ks vs inertias
plt.plot(ks, sse, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('SSE')
plt.xticks(ks)
plt.show()
```

#best k = 8

```python
#Applying kmeans to the dataset with the optimal k value
k = 8
kmeans = KMeans(n_clusters = k)
y_kmeans = kmeans.fit_predict(X)
col_names = list(dff.columns.values)

# append cluster labels to the last column of the dataset
result = pd.concat([pd.DataFrame(dff,columns=col_names), pd.DataFrame(y_kmeans,columns=['ClusterLabels']) ], axis=1)

#cluster label starts from 0
result
```

```python
# Sum of Squared Error of the oprimum model
sse = kmeans.inertia_
print("Sum of Squared Error (SSE): ", sse)
```

```python
# Centriod for each cluster
print("Centroid data:")
centriods_data = mms.inverse_transform(kmeans.cluster_centers_)
centriods_data = pd.DataFrame(centriods_data,columns=col_names)
centriods_data
```
